```python
import pandas as pd #we will use pandas

df = pd.read_csv("Covid Data.csv")

print(df.info()) #Checking the details about our data.

#print(df.head(5)) #print the head to have a feel of what the data looks like

#---------------- CHECKING MISSING VALUES ---------------------------

df.isnull().sum() #sum of the missing values for each attribute

#-------------------REMOVING DUPLICATES------------------------

#df = df.drop_duplicates() # We are not removing any duplicates, since the patients are all anonymous.
#print(df.info())

#----------------FIXING DATATYPES------------------

#print(df.dtypes) #see the datatypes it has

#df['DATE_DIED'] = pd.to_datetime(df['DATE_DIED'], format="%d/%m/%Y")

#date_died = df["DATE_DIED"] #isolate the date row so I can later print
#print(date_died.head(10)) #print the head to better understand what it look like 


#--------------PATIENT SURVIVAL PROCESSING -------------


#df["SURVIVED?"] = df["DATE_DIED"].apply(lambda x: "Yes" if x == "9999-99-99" else "No")
df["SURVIVED?"] = df["DATE_DIED"].apply(lambda x: 1 if x == "9999-99-99" else 0) #1 for survived and 0 for dead 

#print(df.head(10))

#------------------NORMALIZING VALUES---------------------------

x = df.columns.drop("AGE").drop("CLASIFFICATION_FINAL").drop("DATE_DIED").drop("USMER").drop("MEDICAL_UNIT") #we drop age becasue We dont wanna change any of its values
df[x] = df[x].replace(to_replace = 2, value = 0) #this will mean no
df[x] = df[x].replace(to_replace = 1, value = 1) #this is redundant but will mean yes
df[x] = df[x].replace(to_replace = 97, value = -1) #this will mean unknown
df[x] = df[x].replace(to_replace = 98, value = -1) #this will mean unknown
df[x] = df[x].replace(to_replace = 99, value = -1) #this will mean unknown



#print(df.head(10))

#---------------------FILTERING NON COVID----------------------

print((df["CLASIFFICATION_FINAL"] >= 4).sum()) #count the num of confirmend covid and non confirmed covids 
print((df["CLASIFFICATION_FINAL"] < 4).sum())


cleaned_covid_positive_and_negative = df
cleaned_covid_positive_only = df[df["CLASIFFICATION_FINAL"] < 4] #only keep the rows that are < 4, meaning covid confiremed

#----------TO CSV--------------

cleaned_covid_positive_only.to_csv("cleaned_covid_positive_only.csv", index = False)
cleaned_covid_positive_and_negative.to_csv("cleaned_covid_positive_and_negative.csv", index = False)# without index, it created a numbered no name column 

#print((cleaned_covid_positive_only["CLASIFFICATION_FINAL"] >= 4).sum()) #count the num of confirmend covid and non confirmed covids 
#print((cleaned_covid_positive_only["CLASIFFICATION_FINAL"] < 4).sum())
```
